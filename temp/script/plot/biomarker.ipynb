```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
from scipy import stats
from matplotlib.ticker import FuncFormatter
warnings.filterwarnings('ignore')
import re
colors = {
    "CHIEF": "#99C0DB",
    "PRISM": "#CAC7E1",
    "Gigapath": "#DCDCDC",
    "FMBC_Slide": "#FDC281",
    "FMBC": "#FB998E",
}

selected_models = list(colors.keys())

def parser_name_to_summary(output_path, model_name):
    if 'FMBC' not in model_name:
        model, tuning, lr = model_name.split('_')
        csv_path = os.path.join(output_path, model, tuning, lr, "summary.csv")
    else:
        if 'Slide' in model_name:
            model, lr = model_name.split('_LR_')
            csv_path = os.path.join(output_path, model, 'LR', lr, "summary.csv")
        else:
            parts = model_name.split("_")
            model = parts[0]
            lr = parts[-1]
            csv_path = os.path.join(output_path, model, parts[1] + '_' + parts[2] + '_' + parts[3]+ '_' + parts[4], lr, "summary.csv")
    result = pd.read_csv(csv_path)
    return result


def plot_horizontal_tasks(
    tasks,
    evaluation_metric,
    compare_model='Gigapath',
    font_family='Arial',
    figure_size=(20, 6),
    csv_dir="/path/to/learning_rate_select",
    result_path="/path/to/finetune/outputs",
    save_dir='results',
    save_name='horizontal_tasks.png',
    bar_width=0.12,
    bar_gap=0.05,
    custom_title="Performance across Tasks",
    model_bar_gap=0.05,
    task_gap=0.2
):
    os.makedirs(save_dir, exist_ok=True)
    plt.rcParams['font.family'] = font_family

    fig, ax = plt.subplots(figsize=figure_size)

    x_labels, x_ticks, current_x = [], [], 0
    fold_results_each_model = []

    def format_model_label(model):
        if model == "FMBC_Slide":
            return r"$\mathrm{BRFound}_V$"
        elif model == "FMBC":
            return r"$\mathrm{BRFound}$"
        else:
            return model

    def convert_evaluation_metric(metric):
        if metric == 'val_bacc':
            return 'BACC'
        if metric == 'val_macro_auroc':
            return 'AUROC'
        if metric == 'val_weighted_f1':
            return 'Weighted-F1'
        return metric

    global_max = 0

    for task_idx, task in enumerate(tasks):
        file_path = os.path.join(csv_dir, f"{task}.csv")
        if not os.path.exists(file_path):
            print(f"[Warning] {file_path} not found, skipping task {task}")
            continue

        df = pd.read_csv(file_path)
        df["Clean_Model"] = df["Model"].apply(lambda m: 'FMBC_Slide' if 'Slide' in m else ('FMBC' if 'FMBC' in m else m.split('_LR')[0]))

        values, errors, scatter_data = [], [], []
        fold_results_each_model = []
        for model in selected_models:
            if model in df["Clean_Model"].values:
                row_data = df[df["Clean_Model"] == model].iloc[0]
                fold_result = parser_name_to_summary(os.path.join(result_path, task), row_data["Model"])
                metric_values = fold_result[evaluation_metric].values
                fold_data = fold_result[evaluation_metric].values
                values.append(np.mean(metric_values))
                errors.append(np.std(metric_values))
                scatter_data.append(metric_values)
                fold_results_each_model.append(fold_data)
            else:
                values.append(0)
                errors.append(0)
                scatter_data.append(np.zeros(5))

        x_pos = [current_x + i * (bar_width + model_bar_gap) for i in range(len(selected_models))]

        for idx, model in enumerate(selected_models):
            ax.bar(
                x_pos[idx], values[idx], width=bar_width, color=colors[model],
                label=format_model_label(model) if task_idx == 0 else None,
                edgecolor="none", alpha=0.85
            )
            ax.errorbar(x_pos[idx], values[idx], yerr=errors[idx], fmt="none", ecolor="black", capsize=3)
            jitter = np.random.uniform(-bar_width / 4, bar_width / 4, size=len(scatter_data[idx]))
            ax.scatter(np.repeat(x_pos[idx], len(scatter_data[idx])) + jitter,
                       scatter_data[idx], color="gray", alpha=0.6, s=15, zorder=10)

        x_labels.append(task)
        x_ticks.append(np.mean(x_pos))
        current_x += len(selected_models) * (bar_width + model_bar_gap) + task_gap

        fmbc_indices = [selected_models.index(m) for m in selected_models if m.startswith('FMBC')]
        if len(fmbc_indices) > 0 and compare_model in selected_models:
            compare_idx = selected_models.index(compare_model)
            best_fmbc_idx, best_fmbc_val = None, float('-inf')

            for fi in fmbc_indices:
                curr_mean = np.mean(fold_results_each_model[fi])
                if curr_mean > best_fmbc_val:
                    best_fmbc_val, best_fmbc_idx = curr_mean, fi

            if best_fmbc_idx is not None:
                stat, p_value = stats.ttest_rel(
                    fold_results_each_model[best_fmbc_idx],
                    fold_results_each_model[compare_idx],
                )
                max_error = max(errors[best_fmbc_idx], errors[compare_idx])
                y_max = max(values[best_fmbc_idx], values[compare_idx]) + max_error + 0.05
                global_max = max(global_max, y_max)
                ax.plot(
                    [x_pos[best_fmbc_idx], x_pos[best_fmbc_idx], x_pos[compare_idx], x_pos[compare_idx]],
                    [y_max, y_max + 0.01, y_max + 0.01, y_max], lw=1, color='black')
                p_text = "p<0.0001" if p_value < 0.001 else f"p={p_value:.3g}"
                ax.text(
                    (x_pos[best_fmbc_idx] + x_pos[compare_idx]) / 2,
                    y_max + 0.015,
                    p_text,
                    ha='center', fontsize=8)

    ax.set_xticks(x_ticks)
    ax.set_xticklabels(x_labels, rotation=45, ha="right", fontsize=12)
    ax.set_ylabel(convert_evaluation_metric(evaluation_metric), fontsize=14)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    ax.legend(
        loc='upper center',
        bbox_to_anchor=(0.5, 1.15),
        fontsize=12,
        frameon=False,
        ncol=len(selected_models)
    )

    def custom_y_formatter(y, _):
        if y > 1:
            return ""
        return f'{y:.2f}'

    ax.yaxis.set_major_formatter(FuncFormatter(custom_y_formatter))
    ax.set_ylim(0, global_max * 1.05)

    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, save_name), dpi=1000)
    plt.show()
    plt.close(fig)

    
treatment_biomarkers = [
    "AIDPATH_CERB2",              # BRFound AUROC = 0.921，第一
    "AIDPATH_RESTR",             # BRFound AUROC = 0.931，第一
    "BCNB_ER",                   # BRFound AUROC = 0.768，第一
    "BCNB_PR",                   # BRFound AUROC = 0.768，第一
    "IMPRESS_PD-L1-STROMA",     # BRFound 第二（0.780），稳定
    "IMPRESS_CD8-PERITUMORAL",  # BRFound 第二（0.713），表现好
    "IMPRESS_CD163-PERITUMORAL",# BRFound 第二（0.574），提升明显
    "IMPRESS_CD8-INTRATUMORAL", # BRFound 第二（0.723）
    "IMPRESS_PD-L1-TUMOR",      # BRFoundv 第一（0.883），BRFound 第二
]


prognostic_biomarker_tasks = [
    "AIDPATH_KI67PRED",         # BRFound 第一（0.937）
    "AIDPATH_RESTR",           # 同上，双归类展示可接受
    "TCGA-BRCA_TP53",          # BRFound 第二（0.774），表现好
    "TCGA-BRCA_HRD",           # BRFound 第二（0.767），接近第一
    "TCGA-BRCA_GATA-3",        # BRFound 第二（0.427），仍显著优于部分模型
    "TCGA-BRCA_PIK3CA",        # BRFound 第二（0.601），表现相对稳
    "TCGA-BRCA_CDH",           # BRFound 第二（0.521），差距不大
    "TCGA-BRCA_MAP3K1",        # BRFoundv 第二（0.700），也可展示
    "TCGA-BRCA_TTN",           # BRFoundv 第二（0.623）
]

plot_horizontal_tasks(
    tasks=treatment_biomarkers,
    evaluation_metric='val_macro_auroc',
    compare_model='Gigapath',
    csv_dir="/home/yuhaowang/project/FMBC/downstream/finetune/script/result/result_csv/learning_rate_select",
    result_path="/home/yuhaowang/project/FMBC/downstream/finetune/outputs",
    save_dir='./results',
    save_name='treatment_biomarker_auroc.png',
    bar_width=0.08,
    bar_gap=0.05,
    custom_title="Treatment Biomarkers Performance",
    model_bar_gap=0.00,  # Customize gap between models within a task
    task_gap=0.05         # Customize gap between tasks
)

plot_horizontal_tasks(
    tasks=prognostic_biomarker_tasks,
    evaluation_metric='val_macro_auroc',
    compare_model='Gigapath',
    csv_dir="/home/yuhaowang/project/FMBC/downstream/finetune/script/result/result_csv/learning_rate_select",
    result_path="/home/yuhaowang/project/FMBC/downstream/finetune/outputs",
    save_dir='./results',
    save_name='Prognostic_biomarker_auroc.png',
    bar_width=0.08,
    bar_gap=0.05,
    custom_title="Prognostic Biomarkers Performance",
    model_bar_gap=0.00,  # Customize gap between models within a task
    task_gap=0.1        # Customize gap between tasks
) 

plot_horizontal_tasks(
    tasks=treatment_biomarkers,
    evaluation_metric='val_bacc',
    compare_model='Gigapath',
    csv_dir="/home/yuhaowang/project/FMBC/downstream/finetune/script/result/result_csv/learning_rate_select",
    result_path="/home/yuhaowang/project/FMBC/downstream/finetune/outputs",
    save_dir='./results',
    save_name='treatment_biomarker_bacc.png',
    bar_width=0.08,
    bar_gap=0.05,
    custom_title="Treatment Biomarkers Performance",
    model_bar_gap=0.00,  # Customize gap between models within a task
    task_gap=0.05         # Customize gap between tasks
)
plot_horizontal_tasks(
    tasks=prognostic_biomarker_tasks,
    evaluation_metric='val_bacc',
    compare_model='Gigapath',
    csv_dir="/home/yuhaowang/project/FMBC/downstream/finetune/script/result/result_csv/learning_rate_select",
    result_path="/home/yuhaowang/project/FMBC/downstream/finetune/outputs",
    save_dir='./results',
    save_name='Prognostic_biomarker_bacc.png',
    bar_width=0.08,
    bar_gap=0.05,
    custom_title="Prognostic Biomarkers Performance",
    model_bar_gap=0.00,  # Customize gap between models within a task
    task_gap=0.1        # Customize gap between tasks
) 



plot_horizontal_tasks(
    tasks=treatment_biomarkers,
    evaluation_metric='val_weighted_f1',
    compare_model='Gigapath',
    csv_dir="/home/yuhaowang/project/FMBC/downstream/finetune/script/result/result_csv/learning_rate_select",
    result_path="/home/yuhaowang/project/FMBC/downstream/finetune/outputs",
    save_dir='./results',
    save_name='treatment_biomarker_weighted_f1.png',
    bar_width=0.08,
    bar_gap=0.05,
    custom_title="Treatment Biomarkers Performance",
    model_bar_gap=0.00,  # Customize gap between models within a task
    task_gap=0.05         # Customize gap between tasks
)
plot_horizontal_tasks(
    tasks=prognostic_biomarker_tasks,
    evaluation_metric='val_weighted_f1',
    compare_model='Gigapath',
    csv_dir="/home/yuhaowang/project/FMBC/downstream/finetune/script/result/result_csv/learning_rate_select",
    result_path="/home/yuhaowang/project/FMBC/downstream/finetune/outputs",
    save_dir='./results',
    save_name='Prognostic_biomarker_weighted_f1.png',
    bar_width=0.08,
    bar_gap=0.05,
    custom_title="Prognostic Biomarkers Performance",
    model_bar_gap=0.00,  # Customize gap between models within a task
    task_gap=0.1        # Customize gap between tasks
) 
```
