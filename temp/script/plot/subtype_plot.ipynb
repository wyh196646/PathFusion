```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker 
import warnings
from scipy import stats
warnings.filterwarnings('ignore')

colors = {
    "CONCH": "#A3DBD2",
    "CHIEF": "#99C0DB",
    "PRISM": "#CAC7E1",
    "Gigapath": "#DCDCDC",
    "FMBC_Slide": "#FDC281",
    "FMBC": "#FB998E",
}

selected_models = list(colors.keys())
#  subplot title 
task_title_map = {
    "AHSL-NON-IDC-GRADE": "AHSL-NON-IDC",
    "CAMELYON16_TEST_CANCER": "CAMELYON16_CANCER",
    "CAMELYON17_STAGE_4SUBTYPING": "CAMELYON17_STAGE4",
}



def parser_name_to_summary(output_path, model_name):
    """"""
    if 'FMBC' not in model_name:
        model, tuning, lr = model_name.split('_')
        csv_path = os.path.join(output_path, model, tuning, lr, "summary.csv")
    else:
        if 'Slide' in model_name:
            model, lr = model_name.split('_LR_')
            csv_path = os.path.join(output_path, model, 'LR', lr, "summary.csv")
        else:
            parts = model_name.split("_")
            model = parts[0]
            lr = parts[-1]
            csv_path = os.path.join(output_path, model, parts[1] + '_' + parts[2] + '_' + parts[3]+ '_' + parts[4], lr, "summary.csv")
    result = pd.read_csv(csv_path)
    return result

def plot_tasks_grid(
    tasks,
    evaluation_metric,
    compare_model='Gigapath',
    grid_shape=(2, 3),
    font_family='Arial',
    wspace=0.3,
    hspace=0.4,
    tight_layout=True,
    figure_size=(8, 8),
    csv_dir="/path/to/learning_rate_select",
    result_path="/path/to/finetune/outputs",
    save_dir='results',
    save_name='M_N_subplots.png',
    bar_width=0.2,          # 
    bar_gap=0.1,            # 
    p_value_alternative='greater'  # Wilcoxon
):
    """
     MN 
    - bar_gap: 
    - p_value_alternative: Wilcoxon 'two-sided'  'greater'/'less'
    """

    # 
    os.makedirs(save_dir, exist_ok=True)

    # 
    plt.rcParams['font.family'] = font_family

    #  MN 
    M, N = grid_shape
    fig, axes = plt.subplots(M, N, figsize=figure_size)

    #  axes  (11/1)
    if M == 1 and N == 1:
        axes = np.array([[axes]])
    elif M == 1 or N == 1:
        axes = axes.reshape(M, N)

    max_subplots = M * N
    tasks_to_plot = tasks[:max_subplots]

    for i, task in enumerate(tasks_to_plot):
        row = i // N
        col = i % N
        ax = axes[row, col]

        file_path = os.path.join(csv_dir, f"{task}.csv")
        if not os.path.exists(file_path):
            print(f"[Warning] {file_path} not found,  {task}")
            continue

        df = pd.read_csv(file_path)

        # 
        def clean_model_name(model_name):
            model = model_name.split('_LR')[0]
            if model == 'FMBC':
                return 'FMBC'
            elif 'FMBC' in model:
                return 'FMBC_Slide'
            else:
                return model
            
        def format_model_label(model):
            if model == "FMBC_Slide":
                return r"$\mathrm{BRFound}_V$"  #  LaTeX 
            elif model == "FMBC":
                return r"$\mathrm{BRFound}$"
            else:
                return model
            
        def convert_evaluation_metric(metric):
            if metric == 'val_bacc':
                return 'BACC'
            if metric == 'val_macro_auroc':
                return 'MA-AUROC'

        df["Clean_Model"] = df["Model"].apply(clean_model_name)
        df.loc[df["Clean_Model"] == "FMBC_LR_Different_MeanPool", "Clean_Model"] = "FMBC_LR_Same_MeanPool"

        #  std
        values = []
        errors = []
        fold_results_each_model = []

        # 
        for model in selected_models:
            if model in df["Clean_Model"].values:
                row_data = df[df["Clean_Model"] == model].iloc[0]
                n_fold_result = parser_name_to_summary(os.path.join(result_path, task), row_data["Model"])
                fold_data = n_fold_result[evaluation_metric].values
                mean_val = np.mean(fold_data)
                std_val = np.std(fold_data)

                values.append(mean_val)
                errors.append(std_val)
                fold_results_each_model.append(fold_data)
            else:
                #  0 
                values.append(0)
                errors.append(0)
                fold_results_each_model.append(np.array([0] * 5))

        #  x 
        x_positions = np.arange(len(selected_models)) * (bar_width + bar_gap)

        # 
        for idx, model in enumerate(selected_models):
            ax.bar(
                x_positions[idx],
                values[idx],
                width=bar_width,
                color=colors[model],
                edgecolor="none",
                alpha=0.85,
                label=model if i == 0 else None  # 
            )
            ax.errorbar(
                x_positions[idx],
                values[idx],
                yerr=errors[idx],
                ecolor="black",
                capsize=3,
                linewidth=1
            )
            # 
            jitter = np.random.uniform(-bar_width / 4, bar_width / 4, size=len(fold_results_each_model[idx]))
            ax.scatter(
                np.repeat(x_positions[idx], len(fold_results_each_model[idx])) + jitter,
                fold_results_each_model[idx],
                color="gray",
                alpha=0.6,
                s=12,
                zorder=10
            )

        #  Wilcoxon  best FMBC  compare_model
        # T
        fmbc_indices = [selected_models.index(m) for m in selected_models if m.startswith('FMBC')]
        if len(fmbc_indices) > 0 and compare_model in selected_models:
            compare_idx = selected_models.index(compare_model)
            best_fmbc_idx = None
            best_fmbc_val = float('-inf')

            #  FMBC 
            for fi in fmbc_indices:
                curr_mean = np.mean(fold_results_each_model[fi])
                if curr_mean > best_fmbc_val:
                    best_fmbc_val = curr_mean
                    best_fmbc_idx = fi

            if best_fmbc_idx is not None:
                #  P 
                stat, p_value = stats.ttest_rel(
                    fold_results_each_model[best_fmbc_idx],
                    fold_results_each_model[compare_idx],
                )

                #  P 
                max_error = max(errors[best_fmbc_idx], errors[compare_idx])
                y_max = max(values[best_fmbc_idx], values[compare_idx]) + max_error + 0.05
                ax.plot(
                    [x_positions[best_fmbc_idx], x_positions[best_fmbc_idx],
                    x_positions[compare_idx], x_positions[compare_idx]],
                    [y_max, y_max + 0.01, y_max + 0.01, y_max],
                    lw=1, color='black'
                )
                ax.text(
                    (x_positions[best_fmbc_idx] + x_positions[compare_idx]) / 2,
                    y_max + 0.015,
                    f"p={p_value:.3g}",
                    ha='center',
                    fontsize=8
                )

        #  x 
        ax.set_xticks(x_positions)
        ax.set_xticklabels([format_model_label(m) for m in selected_models], rotation=45, ha="right", fontsize=10)

        # 
        plot_task_name = task_title_map.get(task, task)  # 
        ax.set_title(f"{plot_task_name}", fontsize=12, pad=10)
        # y 
        ax.set_ylabel(convert_evaluation_metric(evaluation_metric))
        ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))
        # 
        ax.spines["top"].set_visible(False)
        ax.spines["right"].set_visible(False)

    #  handles & labels 
    handles, labels = axes[0, 0].get_legend_handles_labels()

    # 
    if tight_layout:
        plt.tight_layout()
    else:
        plt.subplots_adjust(wspace=wspace, hspace=hspace)

    # 
    plt.savefig(os.path.join(save_dir, save_name), dpi=300)
    plt.show()
    plt.close(fig)

# ---------------------------
# /
# ---------------------------
total_task_list=os.listdir('/home/yuhaowang/project/FMBC/downstream/finetune/dataset_csv/subtype')

total_task_list=[task[:-4] for task in total_task_list if task[-4:]=='.csv']
trash_task = ['DORID_2','POST-NAT-BRCA-3TYPE','BACH_TUMOR','AHSL-SUBTYPE','CPTAC-IDC']
#AHSL-SUBTYPE 
not_sota_task = ['CAMELYON16_TEST_IDC','BCNB_TUMOR','IMPRESS_RESIDUAL-TUMOR','BRACS_FINE','AHSL-GRADE-1','TUPAC_TUMOR_SCORE',
                 'AIDPATH_GRADE','CPTAC_IDC','CPTAC_AJCC8SUBTYPE']#'AHSL_SUBTYPE',
not_good_task = ['POST-NAT-BRCA-HERIHC']
task_list = [task for task in total_task_list if task not in trash_task and task not in not_sota_task and task not in not_good_task]

print('The number of sota tasks is {0},across {1} datasets'.format(len(task_list),len(set([task.split('_')[0] for task in task_list]))))
    
if __name__ == "__main__":
    example_tasks = task_list
    print(task_list)

Subtyping = [
    "TCGA-BRCA-SUBTYPE",
    "AIDPATH_IDC",
    "AHSL_SUBTYPE",
    "BRACS_COARSE",
    "DORID_6",
    "BRACS_FINE",
    "CAMELYON16_TEST_CANCER",
    "CAMELYON17_STAGE_4SUBTYPING",
]

grading = [
"SLNBREAST_SUBTYPE",
"AHSL-NON-IDC-GRADE",
"TCGA-BRCA_M",
"TCGA-BRCA_N",
"TCGA-BRCA_T",
"BCNB_ALN",
"TCGA-BRCA_STAGE",
"CAMELYON17_STAGE"
]
plot_tasks_grid(
    tasks=Subtyping,
    evaluation_metric='val_bacc',#val_macro_auroc,val_bacc
    compare_model='Gigapath',
    grid_shape=(2, 4),
    font_family='Arial',
    wspace=0.4,
    hspace=0.4,
    tight_layout=True,
    figure_size=(16,8),
    csv_dir="/home/yuhaowang/project/FMBC/downstream/finetune/script/result/result_csv/learning_rate_select",
    result_path="/home/yuhaowang/project/FMBC/downstream/finetune/outputs",
    save_dir='./results',
    save_name='subtype_subplots_bacc.png',
    bar_width=0.1,               # 
    bar_gap=0.05,                 # 
    p_value_alternative='greater'  # 
)
plot_tasks_grid(
    tasks=grading,
    evaluation_metric='val_bacc',
    compare_model='Gigapath',
    grid_shape=(2, 4),
    font_family='Arial',
    wspace=0.4,
    hspace=0.4,
    tight_layout=True,
    figure_size=(16,8),
    csv_dir="/home/yuhaowang/project/FMBC/downstream/finetune/script/result/result_csv/learning_rate_select",
    result_path="/home/yuhaowang/project/FMBC/downstream/finetune/outputs",
    save_dir='./results',
    save_name='grading_subplots_bacc.png',
    bar_width=0.1,              
    bar_gap=0.05,                 
    p_value_alternative='greater' 
    
)



plot_tasks_grid(
    tasks=Subtyping,
    evaluation_metric='val_macro_auroc',#val_macro_auroc,val_bacc
    compare_model='Gigapath',
    grid_shape=(2, 4),
    font_family='Arial',
    wspace=0.4,
    hspace=0.4,
    tight_layout=True,
    figure_size=(16,8),
    csv_dir="/home/yuhaowang/project/FMBC/downstream/finetune/script/result/result_csv/learning_rate_select",
    result_path="/home/yuhaowang/project/FMBC/downstream/finetune/outputs",
    save_dir='./results',
    save_name='subtype_subplots_val_macro_auroc.png',
    bar_width=0.1,               # 
    bar_gap=0.05,                 # 
    p_value_alternative='greater'  # 
)
plot_tasks_grid(
    tasks=grading,
    evaluation_metric='val_macro_auroc',
    compare_model='Gigapath',
    grid_shape=(2, 4),
    font_family='Arial',
    wspace=0.4,
    hspace=0.4,
    tight_layout=True,
    figure_size=(16,8),
    csv_dir="/home/yuhaowang/project/FMBC/downstream/finetune/script/result/result_csv/learning_rate_select",
    result_path="/home/yuhaowang/project/FMBC/downstream/finetune/outputs",
    save_dir='./results',
    save_name='grading_subplots_val_macro_auroc.png',
    bar_width=0.1,              
    bar_gap=0.05,                 
    p_value_alternative='greater' 
    
)
```
