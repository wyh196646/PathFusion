```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

def normalize_csv(file_path, output_path):
    # 读取CSV文件
    df = pd.read_csv(file_path)
    
    # 确保'Slide ID'列存在
    if 'slide_id' not in df.columns:
        raise ValueError("CSV 文件中缺少 'slide id' 列")
    
    # 复制 'slide id' 列
    slide_id = df[['slide_id']]
    
    # 选择需要归一化的列（除了 'slide id' 之外）
    data_columns = df.drop(columns=['slide_id'])
    
    # 进行 Min-Max 归一化
    scaler = MinMaxScaler()
    normalized_data = scaler.fit_transform(data_columns)
    
    # 转换回 DataFrame
    normalized_df = pd.DataFrame(normalized_data, columns=data_columns.columns)
    
    # 重新组合数据
    result_df = pd.concat([slide_id, normalized_df], axis=1)
    
    # 保存归一化后的数据到新的 CSV 文件
    result_df.to_csv(output_path, index=False)
    
    print(f"归一化后的数据已保存到 {output_path}")

data_path='/home/yuhaowang/project/FMBC/downstream/finetune/dataset_csv/gene_exp/CPTAC_PROTEIN.csv'   
normalize_csv(data_path,data_path)
```
